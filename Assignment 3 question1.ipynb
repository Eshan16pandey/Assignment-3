```python
import pandas as pd
from scipy import stats
import numpy as np
```

```python
df = pd.read_csv('/Users/eshanakshata/Desktop/Cutlets.csv')
```

```python
df
```

```python
Unit_a = df['Unit A']
Unit_b = df['Unit B']
```

```python
t_stat, p_value = stats.ttest_ind(Unit_a,Unit_b)

```

```python
#significance level

alpha = 0.05
```

```python
print("t-statistic:",t_stat)
print("P value:",p_value)

```

```python
if p_value < alpha:
    print("reject the null hypothesis :there is a significant difference in the diameter of cutlet between Unit A and Unit B.")
else:
    print("Fail to reject null hypothesis: there is no significant difference in the diameter of cutlet between Unit A and Unit B.")
    
```

```python
#Based on the t-statistic of approximately 0.723 and the p-value of approximately 0.472, 
#we fail to reject the null hypothesis. 
#This means that there is no significant difference in the diameter of the cutlet between Unit A and Unit B. 
#The assumptions of normality and equality of variances appear to be met based on the tests performed.
```
